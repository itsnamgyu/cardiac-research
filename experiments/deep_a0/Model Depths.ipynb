```python
from core.fine_model import FineModel
```

```python
models = {}
for key, Model in FineModel.get_dict().items():
    models[key] = Model()
    models[key].get_model()
```

```python
for key, fm in models.items():
    print((" " + key + " ").center(80, "#"))
    for index, layer in enumerate(fm.get_model().layers):
        print(index, layer.name)
```
