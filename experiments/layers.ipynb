```python
for app in ku.apps.values():
    print(app.name.center(80, '-'))
    count = 0
    for layer in app.get_model().layers:
        if type(layer) is keras.layers.convolutional.Conv2D:
            count += 1
    print('Number of Conv Layers: {}'.format(count))
```

```python
resnet = ku.apps['resnet50'].get_model()
```

```python
for index, layer in enumerate(resnet.layers):
    print('{:02d}. {}'.format(index, str(layer)))
    if type(layer) is keras.layers.convolutional.Conv2D:
        print('Conv Layer'.center(80, '-'))
        print('Kernel Size: {}'.format(layer.kernel_size))
        print('Filters: {}'.format(layer.filters))
        print(''.center(80, '-'))
        index += 1
```
