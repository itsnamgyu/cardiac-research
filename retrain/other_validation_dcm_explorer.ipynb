```python
import os
import glob

import dcm_utils as dcmu
import pydicom
import matplotlib.pyplot as plt
from natsort import natsorted
```

```python
path_format = 'validate/501/study/sax_26/IM-11494-00%02d.dcm'
dcms = []
plt.figure(figsize=(15,10))
for i in range(1, 30):
    path = path_format % i
    dcms.append(pydicom.dcmread(path))
    plt.subplot(3, 10, i)
    dcmu.plot_dicom(dcms[i - 1])
```

```python
paths = glob.glob('validate/504/study/sax_*/IM-*-0001.dcm')
plt.figure(figsize=(15,10))

paths = natsorted(paths)
for i, path in enumerate(paths):
    plt.subplot(3, 10, i + 1)
    plt.title(path[20:25])
    dcmu.plot_dicom(pydicom.dcmread(path))
```
